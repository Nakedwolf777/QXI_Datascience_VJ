```python
mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
x = np.linspace(0,7,10)
y = np.sin(x)
plt.plot(x,y,'go')
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
```

```python
plt.figure(figsize=(10,5))
x = np.linspace(0,7,10)
y = np.sin(x)
plt.plot(x,y,'r-')
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
```

```python
plt.figure(figsize=(7,5))
x = np.arange(1,8)
y1 = np.power(x,2)
y2 = np.power(x,3)
plt.plot(x,y1, 'g-', label = '$y1 = x^2&') 
plt.plot(x,y2, 'b-', label = '$y2 = x^3$')
plt.legend
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.tight_layout
plt.legend(loc = "upper right")
```

```python
plt.figure(figsize=(16,9))
x = np.linspace(0,10,1000)
plt.plot(x,np.sin(x), label = '$Sin(X)$  $Dashed$', linestyle = '--' )
plt.plot(x+1,np.sin(x), label = '$Sin(X)$  $Dashdot$', linestyle = '-.')
plt.plot(x,np.cos(x), label = '$Cos(X)$  $Solid$', linestyle = '-' )
plt.plot(x+1,np.cos(x), label = '$Cos(X)$  $Dotted$', linestyle = ':')
plt.title('$Sin(X)$  $&$  $Cos(X)$', fontsize=14)
plt.legend(loc='upper right', fontsize = 14, bbox_to_anchor = (1.2,1))

         
```
